## ***IMPORTING LIBRARIES***

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
```

## ***LOADING THE DATASET***

```python
dataset = load_digits()
dataset
```

## ***SUMMARIZING THE DATASET***

```python
print(dataset.images.shape)
print(dataset.images)
print(dataset.target.shape)
print(dataset.target)
```

## ***VISUALIZING THE DATASET***

```python
import matplotlib.pyplot as plt

n = 789 # sample number out of 1797 total sample

print(dataset.images[n])

plt.gray()
plt.imshow(dataset.images[n])
```

## ***SEGREGATING THE DATASET INTO INPUT AND OUTPUT***

```python
input  = dataset.images.reshape(len(dataset.images),-1)
print(input)
output = dataset.target
print(output)
```

## ***SPLITTING THE DATASET INTO TRAIN AND TEST DATA***

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(input,output)
```

## ***TRAINING THE DATASET***

```python
from sklearn import svm

SVM_SVC_Model = svm.SVC(kernel="linear")
SVM_SVC_Model.fit(x_train,y_train)
```

## ***PREDICTING THE RESULT USING THE TRAINED MODEL***

```python
y_pred = SVM_SVC_Model.predict(x_test)
y_pred
```

## ***CALCULATING THE ACCURACY OF THE TRAINED MODEL***

```python
from sklearn.metrics import accuracy_score
print("accuracy_score",accuracy_score(y_pred,y_test)*100,"%")
```

## ***PREDICTING THE OUTPUT OF SINGLE TEST DATA USING THE TRAINED MODEL***

```python
print("ACUTAL OUTPUT: ",y_test[55])
plt.imshow(np.reshape(x_test[55],(8,8)))
print("PREDICTED OUTPUT: ")
SVM_SVC_Model.predict(np.reshape(x_test[55],(1,64)))
```
